```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools = len(school_data["school_name"].value_counts())
```

```python
# Calculate the total number of students
total_students = len(student_data["student_name"].value_counts())
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
```

```python
# Calculate the average math score
average_math = student_data["math_score"].mean()
```

```python
# Calculate the average reading score
average_reading = student_data["reading_score"].mean()
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
students_passed_math = (student_data["math_score"].loc[student_data["math_score"] >= 70]).count() / student_data["student_name"].count() * 100
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
students_passed_reading = (student_data["reading_score"].loc[student_data["reading_score"] >= 70]).count() / student_data["student_name"].count() * 100
```

```python
# Calculate percentage of students who passed math & reading (%Overall Passing)
passing_math_reading_count = school_data_complete[(school_data_complete["math_score"] >= 70) 
                                                & (school_data_complete["reading_score"] >= 70)].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(total_students) * 100
```

```python
# Create a new dataframe
raw_data = {
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math],
    "Average Reading Score": [average_reading],
    "Percent Math Passing": [students_passed_math],
    "Percent Reading Passing": [students_passed_reading],
    "Percent Overall Passing": [overall_passing_rate]
    }

new_data_df = pd.DataFrame(raw_data, columns=[
                        "Total Schools",
                        "Total Students", 
                        "Average Math Score", 
                        "Average Reading Score", 
                        "Percent Math Passing", 
                        "Percent Reading Passing", 
                        "Percent Overall Passing"])
new_data_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#it will be helpful to set_index to (["school_name"])["type"]
#to get these calculations you are going to use a number of methods

# Determine school type
school_type = school_data.set_index("school_name")["type"]
```

```python
# Calculate the total student count per school
students_per_school = school_data_complete.groupby("school_name")["Student ID"].count()
```

```python
# Calculate the school budget per school
total_school_budget = school_data.groupby("school_name")["budget"].sum()
```

```python
# Calculate the per student budget
total_student_budget = total_school_budget / students_per_school
```

```python
# Average Math Score & Reading Score
total_average_math = school_data_complete.groupby("school_name")["math_score"].mean()
total_average_reading = school_data_complete.groupby("school_name")["reading_score"].mean()
```

```python
# Percent Passing Math
school_passing_math = (school_data_complete[school_data_complete["math_score"] >= 70])
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / students_per_school * 100

# Percent Passing Reading
school_passing_reading = (school_data_complete[school_data_complete["reading_score"] >= 70])
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / students_per_school * 100
```

```python
# Percent Overall Passing
school_overall_passing = (per_school_passing_math + per_school_passing_reading) / 2
```

```python
# Create dataframe
school_summary_df = pd.DataFrame({"School Type": school_type,
      "Total Students": students_per_school,
      "Total School Budget": total_school_budget,
      "Per Student Budget": total_student_budget,
      "Average Math Score": total_average_math,
      "Average Reading Score": total_average_reading,
      "% Passing Math": per_school_passing_math,
      "% Passing Reading": per_school_passing_reading,
      "% Overall Passing Rate": school_overall_passing})
school_summary_df
```
